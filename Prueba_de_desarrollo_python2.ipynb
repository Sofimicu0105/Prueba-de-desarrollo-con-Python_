```python
import numpy as np
import matplotlib.pyplot as plt

# Genera un rango de valores para x
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Calcula los valores correspondientes de la función seno
y = np.sin(x)

# Encuentra los índices de los puntos máximos y mínimos
max_indices = np.where(np.diff(np.sign(np.diff(y))) < 0)[0] + 1
min_indices = np.where(np.diff(np.sign(np.diff(y))) > 0)[0] + 1

# Imprime las coordenadas de los puntos máximos y mínimos
print("Puntos máximos:")
for i in max_indices:
    print(f"x = {x[i]:.2f}, y = {y[i]:.2f}")

print("\nPuntos mínimos:")
for i in min_indices:
    print(f"x = {x[i]:.2f}, y = {y[i]:.2f}")

# Grafica la función seno y resalta los puntos máximos y mínimos
plt.plot(x, y)
plt.plot(x[max_indices], y[max_indices], "ro", label="Máximos")
plt.plot(x[min_indices], y[min_indices], "bo", label="Mínimos")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Puntos máximos y mínimos de la función seno")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 32, 500)
# Calcular valores y como seno de x
y = np.sin(x)

# Encontrar índices de los puntos máximos y mínimos
max_indices = np.where((y[1:-1] > y[:-2]) & (y[1:-1] > y[2:]))[0] + 1
min_indices = np.where((y[1:-1] < y[:-2]) & (y[1:-1] < y[2:]))[0] + 1
x_max = x[max_indices]
y_max = y[max_indices]
x_min = x[min_indices]
y_min = y[min_indices]
plt.plot(x, y, label='sin(x)')
plt.scatter(x_max, y_max, color='red', label='Máximos')
plt.scatter(x_min, y_min, color='blue', label='Mínimos')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Seno con puntos máximos y mínimos')
plt.legend()
plt.grid(True)
plt.show()
```
